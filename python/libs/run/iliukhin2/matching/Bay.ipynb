```python
def is_PowerTransformer(text):
    if isinstance(text, str):
        match = re.findall('^[ТT]-\d', text)
        if match:
            res = re.findall('\d', match[0])
            return res[0]
        match = re.findall('^\dТ', text)
        if match:
            res = re.findall('\d', match[0])
            return res[0]
    return None

df_rs['is_PowerTransformer'] = df_rs['name'].apply(is_PowerTransformer)
# df_rs[~df_rs['is_PowerTransformer'].isna()]['name'].unique()
```

```python
def is_Busbar(text):
    if isinstance(text, str):
        match1 = re.findall('с\.?ш\.?-? ?\d', text.lower())
        if match1:
            res = re.findall('\d', match1[0])
            return f'сш{res[0]}'
            
        match2 = re.findall('[\dО]сш', text.lower())
        if match2:
            res = re.findall('\d', match2[0])
            return f'сш{res[0]}'
    return None

df_rs['is_Busbar'] = df_rs['name'].apply(is_Busbar)
# display(df_rs[~df_rs['is_Busbar'].isna()]['name'].unique())
# display(df_rs[~df_rs['is_Busbar'].isna()]['is_Busbar'].unique())

df_supa['is_Busbar'] = df_supa['Название'].apply(is_Busbar)
# df_supa['is_Busbar'] = df_supa['Дисп_наименование'].apply(is_busbar)
# display(df_supa[~df_supa['is_Busbar'].isna()]['Название'].unique())
```
