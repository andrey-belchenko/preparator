```python
import pandas as pd
import numpy as np
import re
from collections import Counter

pd.set_option('max_colwidth', 100)
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
def take_needed_cols(path, cols_from_modeller, cols2preparator, id2check=None):
    df = pd.read_csv(path, sep=';')
    df = df[~df['_id|R-Equ'].isnull()]
    
    assert len(cols_from_modeller) == len(cols2preparator)
    mod2prep = {key:val for key, val in zip(cols_from_modeller, cols2preparator)}
    df = df[cols_from_modeller]
    df = df.rename(columns=mod2prep)
    # df = df.drop_duplicates()
    if id2check:
        assert df[id2check].duplicated().sum() == 0, 'ther are some duplicates in ids'
    return df

def clear_iri(text):
    match = re.findall('#.*', text)
    if match:
        res = match[0].replace('"', '').replace('#', '')
        return res
```

```python
path = 'data_from_modeller/PowerTransformerQuery.2023-12-26.17_04_06.csv'
df_rs = pd.read_csv(path, sep=';')
cols_from_modeller = ['_id|R', 'IdentifiedObject.name|R', 
                      'class|R', '_id|R-Equ', 'IdentifiedObject.name|R-Equ',
                      ]
cols2preparator = ['Uid', 'name', 'Класс', 'Substation_Uid','Substation_name']

df_rs = take_needed_cols(path, cols_from_modeller, cols2preparator, id2check='Uid')
display(df_rs.head())
df_rs.info()
for col in df_rs.columns:
    print(col, df_rs[col].duplicated().sum())

df_rs.to_csv('output_data/PowerTransfromer/PowerTransformer4preparator_full.csv', index=False)
```
